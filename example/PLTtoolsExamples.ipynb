## Dummy dataset

```python
from sklearn import datasets
```

```python
dummy_dict = dict(
    n_samples = 10000,
    factor = 0.1,
    noise = 0.5,
    random_state = 73,
)
```

```python
X, y = datasets.make_circles(**dummy_dict)
```

```python
def test(X,Y=None):
    if isinstance(Y, (list, np.ndarray)):
        print("2D")
    elif not Y:
        print("1D")
```

```python
test(X=X[:,0])
```

## Figs & Axes

```python
fig_dict = dict(
    P = 3,
    max_col = 2,
    fig_frame = (5,4),
    res = 100,
)
```

```python
fig, ax = PLTtools.get_axes(**fig_dict)
ax[0].scatter(*X.T)
ax[1].scatter(*X.T)
PLTtools.axarrows(axes=ax[1], figure=fig)
```
